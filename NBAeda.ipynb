```python
import pandas as pd
```

```python
draft = pd.read_csv('nbaplayersdraft.csv')

#totalMins = draft['minutes_played'].groupby(['college']).sum()
#draft['minutes_played']
df1 =draft.groupby(['college']).sum()
df1.sort_values(['minutes_played'], ascending=False)


#create df with schools ranked by total minutes played
#only include schools that have players that have played games
df2 = df1[df1['games'] >0].sort_values(['minutes_played'], ascending=False).reset_index()

#simplfied df with schools and their rankings
CollegeRanking_MP =df2[["college",'minutes_played']].reset_index()


CollegeRanking_MP.columns=["college_rank","college","minutes_played"]

CollegeRanking_MP
```

```python
#DF with top 5 schools and their players
#Colleges_top5 = CollegeRanking_MP.head()
Colleges_top5 = draft[draft['college'].isin(['Duke','UNC', 'Kentucky', 'Arizona','UCLA'])]

Colleges_top5

#Average total minutes played (career wise not per game)
Colleges_top5.groupby(['college'])['minutes_played'].mean()

Colleges_top5.groupby(['college'])['points_per_game'].mean()
```
